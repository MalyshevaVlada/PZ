```python
import sqlite3 as sq
```

***СОЗДАНИЕ ТАБЛИЦ***

```python
with sq.connect('deanery.db') as con:
    cur = con.cursor()
    
    # таблица факультета
    cur.execute("""CREATE TABLE IF NOT EXISTS faculty 
      (
        id_faculty INT auto_increment PRIMARY KEY NOT NULL,
        name_faculty VARCHAR NOT NULL
      );
      """)
    
    # таблица кафедры
    cur.execute("""CREATE TABLE IF NOT EXISTS department
      (
        id_department INT auto_increment PRIMARY KEY NOT NULL,
        name_department VARCHAR NOT NULL,
        id_faculty VARCHAR NOT NULL,
        FOREIGN KEY (id_faculty) REFERENCES faculty(id_faculty) ON UPDATE CASCADE
      );
      """)
    
    # таблица специальности
    cur.execute("""CREATE TABLE IF NOT EXISTS speciality 
    (
      id_speciality INT auto_increment PRIMARY KEY NOT NULL,
      name_speciality VARCHAR NOT NULL,
      id_department INT NOT NULL,
      FOREIGN KEY (id_department) REFERENCES department(id_departament) ON UPDATE CASCADE
    );
    """)

    # таблица предметы
    cur.execute("""CREATE TABLE IF NOT EXISTS subject
    (
      id_subject INT auto_increment PRIMARY KEY NOT NULL,
      name_subject VARCHAR NOT NULL
    );
    """)

    # таблица форма сдачи предмета
    cur.execute("""CREATE TABLE IF NOT EXISTS exam_form 
    (
      id_exam_form INT auto_increment PRIMARY KEY NOT NULL,
      name_exam_form VARCHAR NOT NULL
    );
    """)

    # таблица учебный план
    cur.execute("""CREATE TABLE IF NOT EXISTS syllabus
    (
      id_syllabus INT auto_increment PRIMARY KEY NOT NULL,
      id_speciality INT NOT NULL,
      id_subject INT NOT NULL,
      id_exam_form INT NOT NULL,
      count_of_lecture_hours INT DEFAULT 0,
      count_of_practical_hours INT DEFAULT 0,
      count_of_laboratory_hours INT DEFAULT 0,
      course_work BOOL DEFAULT 0,
      FOREIGN KEY (id_speciality) REFERENCES speciality(id_speciality) ON UPDATE CASCADE,
      FOREIGN KEY (id_subject) REFERENCES subject(id_subject) ON UPDATE CASCADE,
      FOREIGN KEY (id_exam_form) REFERENCES exam_form(id_exam_form) ON UPDATE CASCADE
    );
    """)

    # таблица абитуриенты
    cur.execute("""CREATE TABLE IF NOT EXISTS applicants
    (
      id_applicant INT auto_increment PRIMARY KEY NOT NULL,
      lastname_applicant TEXT NOT NULL,
      firstname_applicant TEXT NOT NULL,
      otchestvo_applicant TEXT,
      gender_applicant VARCHAR(1) CHECK(gender_applicant == 'm' OR gender_applicant == 'f'),
      birthday_applicant DATE NOT NULL,
      address_applicant TEXT CHECK(address_applicant !=''),
      email_applicant TEXT UNIQUE CHECK(email_applicant !=''),
      phone_applicant TEXT UNIQUE CHECK(phone_applicant !=''),
      receipt_date_applicant DATE NOT NULL,
      name_speciality VARCHAR NOT NULL,
      FOREIGN KEY (name_speciality) REFERENCES speciality(name_speciality) ON UPDATE CASCADE
    );""")

    # учебная карточка
    cur.execute("""CREATE TABLE IF NOT EXISTS card
    (
      id_card INT auto_increment PRIMARY KEY NOT NULL,
      lastname_applicant TEXT NOT NULL,
      firstname_applicant TEXT NOT NULL,
      otchestvo_applicant TEXT,
      group_number TEXT NOT NULL,
      name_speciality VARCHAR NOT NULL,
      name_subject VARCHAR NOT NULL,
      name_exam_form VARCHAR NOT NULL,
      rating INT,
      FOREIGN KEY (lastname_applicant) REFERENCES applicant(lastname_applicant) ON UPDATE CASCADE,
      FOREIGN KEY (firstname_applicant) REFERENCES applicant(firstname_applicant) ON UPDATE CASCADE,
      FOREIGN KEY (otchestvo_applicant) REFERENCES applicant(otchestvo_applicant) ON UPDATE CASCADE,
      FOREIGN KEY (name_speciality) REFERENCES speciality(name_speciality) ON UPDATE CASCADE,
      FOREIGN KEY (name_subject) REFERENCES subject(name_subject) ON UPDATE CASCADE,
      FOREIGN KEY (name_exam_form) REFERENCES exam_form(name_exam_form) ON UPDATE CASCADE
    )"""
    )
```

```python
with sq.connect('deanery.db') as con:
    cur = con.cursor()

    # 1. Вывести список всех студентов, зачисленных на факультет, с указанием номера
    # группы.
    cur.execute("""SELECT lastname_applicant, firstname_applicant, otchestvo_applicant, group_number
                    FROM card
                    JOIN speciality ON speciality.name_speciality = card.name_speciality
                    JOIN department ON department.id_department = speciality.id_department
                    JOIN faculty ON faculty.id_faculty = department.id_faculty
                    WHERE faculty.name_faculty = 'Авиастроение'
                """)
    result_1 = cur.fetchall()


    # 2. Вывести список всех специальностей факультета и количество студентов,
    # обучающихся по каждой из них.
    cur.execute("""SELECT speciality.name_speciality, COUNT(card.id_card) student_count
                    FROM faculty
                    INNER JOIN department ON faculty.id_faculty = department.id_faculty
                    INNER JOIN speciality ON department.id_department = speciality.id_department
                    LEFT JOIN card ON speciality.name_speciality = card.name_speciality
                    GROUP BY speciality.name_speciality;
                """)
    result_2 = cur.fetchall()


    # 3. Вывести список всех кафедр факультета и количество студентов, обучающихся на
    # каждой кафедре.
    cur.execute("""SELECT department.name_department, COUNT(card.id_card) as student_count
                    FROM department
                    INNER JOIN speciality ON department.id_department = speciality.id_department
                    LEFT JOIN card ON speciality.name_speciality = card.name_speciality
                    GROUP BY department.name_department;
                """)
    result_3 = cur.fetchall()


    # 4. Вывести список всех предметов и количество часов, выделенных на каждый
    # предмет в учебном плане каждой специальности.
    cur.execute("""SELECT s.name_subject, SUM(sl.count_of_lecture_hours + sl.count_of_practical_hours + sl.count_of_laboratory_hours) AS total_hours
                     FROM subject s
                     INNER JOIN syllabus sl ON s.id_subject = sl.id_subject
                     GROUP BY s.name_subject
                """)
    result_4 = cur.fetchall()


    # 5. Вывести список всех студентов, у которых есть неудовлетворительные оценки
    # (меньше 4) по любому предмету.
    cur.execute("""SELECT card.lastname_applicant, card.firstname_applicant, card.otchestvo_applicant, card.rating
                   FROM card
                   WHERE card.rating < 4
                """)
    result_5 = cur.fetchall()
    

    # 6. Вывести список всех предметов, которые изучают студенты
    cur.execute("""SELECT DISTINCT card.name_subject FROM card""")
    result_6 = cur.fetchall()


    # 7. Вывести список всех студентов, которые сдают курсовую работу
    cur.execute("""SELECT lastname_applicant, firstname_applicant, group_number
                    FROM card
                    INNER JOIN syllabus ON card.name_speciality = syllabus.id_speciality
                    WHERE syllabus.course_work = 1""")
    result_7 = cur.fetchall()


    # 8. Вывести список всех абитуриентов, на специальности "Фармакология и фармация"
    cur.execute("""SELECT lastname_applicant, firstname_applicant, name_speciality 
                    FROM applicants 
                    WHERE name_speciality = "Фармакология и фармация"
                """)
    result_8 = cur.fetchall()


    # 9. Вывести список всех предметов, которые изучают студенты группы 'G-07'. (по вашему проектированию бд у нас один предмет)
    cur.execute("""SELECT card.name_subject FROM card WHERE card.group_number = 'G-07'""")
    result_9 = cur.fetchall()


    # 10. Вывести список студентов и их оценки за все предметы на специальности
    # 'Техническая эксплуатация летательных аппаратов и наземного оборудования' (по вашему проектированию бд у нас одна оценка)
    cur.execute("""SELECT card.lastname_applicant, card.firstname_applicant, card.otchestvo_applicant, card.rating
                    FROM card
                    WHERE card.name_speciality = 'Техническая эксплуатация летательных аппаратов и наземного оборудования'
                """)
    result_10 = cur.fetchall()
```
